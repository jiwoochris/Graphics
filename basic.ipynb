MBTI와 키로 데이터 분석 연습하기

```python
import pandas as pd

# Load the data from the uploaded CSV file
file_path = 'data/Gen.G 아카데미 데이터분석을 위한 설문조사(응답) - 설문지 응답 시트1.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
# 기술 통계 요약
descriptive_stats = data.describe()

descriptive_stats
```

```python
# MBTI 유형 조합 생성
data['MBTI'] = data['여러분들의 MBTI는 뭔가요? (1)'] + data['여러분들의 MBTI는 뭔가요? (2)'] + \
               data['여러분들의 MBTI는 뭔가요? (3)'] + data['여러분들의 MBTI는 뭔가요? (4)']

# MBTI 유형별 빈도수 계산
mbti_counts = data['MBTI'].value_counts()

mbti_counts
```

막대 그래프 그려보기

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Combining the MBTI components into one column for easier analysis
data['MBTI'] = data[['여러분들의 MBTI는 뭔가요? (1)', '여러분들의 MBTI는 뭔가요? (2)', 
                     '여러분들의 MBTI는 뭔가요? (3)', '여러분들의 MBTI는 뭔가요? (4)']].agg(''.join, axis=1)

# Plotting MBTI distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='MBTI', data=data, order = data['MBTI'].value_counts().index)
plt.title('MBTI Type Distribution')
plt.xlabel('MBTI Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# MBTI 첫 번째 요소인 'E' 대 'I'의 분포를 막대그래프로 표현하기

plt.figure(figsize=(6, 4))
sns.countplot(x='여러분들의 MBTI는 뭔가요? (1)', data=data)
plt.title('Distribution of MBTI Component: E vs I')
plt.xlabel('E/I')
plt.ylabel('Count')
plt.show()
```
